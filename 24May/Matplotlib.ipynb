```python
a = pd.read_csv('Book1.csv')

dfs = a.iloc[0:10,0:2]
```

```python
a.info()
```

```python
x = dfs['class']
y = dfs['percentage']
```

```python
#Line plot
```

```python
plt.plot(x,y,'o-.r',ms=12,lw=5)
plt.ylabel("Marks")
plt.xlabel("Students")
plt.title("ADIT Marks",{'color':'cyan', 'size':18,'family':'serif'})
plt.grid(axis='y' , color ='green' , linestyle='-.')
```

```python
# Bar Plot
```

```python
plt.figure(figsize=(15,5))
plt.bar(x,y,color='#A0c0d0')
plt.ylabel("Marks")
plt.xlabel("Students")
plt.title("ADIT Marks",{'color':'cyan', 'size':20,'family':'serif'})
plt.grid( color ='green' , linestyle='-')
```
