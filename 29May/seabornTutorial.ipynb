```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n =sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(2)
```

```python
sns.scatterplot(x='total_bill', y='tip' , data = tips , hue='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='time',sizes=(10,1000),palette='Accent_r')
#time
#style='size'

plt.xlabel('Total Bill in USD')
plt.ylabel('Tips to the service')
plt.title('New York')
```

```python
sns.lineplot(data=tips , x='total_bill' , y='size', hue='sex',style='smoker', markers=True , palette='GnBu') 
```

```python
sns.barplot(data = tips , x='size' , y='total_bill' ,hue='sex',ci=None)

#

```

```python
sns.histplot(x='size',data=tips,bins=5,kde=True)
```

```python
sns.kdeplot(x='size',data=tips,hue='smoker',fill=True,linewidth=20)
```

```python
sns.boxplot(data = tips,x='day',y='tip' , hue='smoker' , linewidth=1)
```

```python
sns.violinplot(data = tips,x='day',y='tip' , hue='sex' ,linewidth=1)
```
