```python
brands =df['Brands']
years =df['2016']

# Plotting
plt.figure(figsize=(4, 4))
plt.pie(years, labels=brands, autopct='%1.1f%%', startangle=140)
plt.title('Units Sold by Brands in 2016')
plt.axis('equal') 
plt.show()
```
