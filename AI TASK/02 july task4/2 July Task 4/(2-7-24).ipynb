## Use iris flower dataset from sklearn library and use cross_val_score against following
models to measure the performance of each. In the end figure out the model with best performance,
# 1. Logistic Regression
# 2. SVM
# 3. Decision Tree
# 4. Random Forest


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data , columns=iris.feature_names)
df['Flower'] = iris.target
df['Flower']=df['Flower'].apply(lambda x : iris.target_names[x])
```

```python
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train ,  x_test ,y_train ,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import svm
```

```python
model = svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvs = cross_val_score
cvs
```

```python
cvs(svm.SVC(kernel='linear' , C=10, gamma='auto'),x_train, y_train,cv=5)
```

```python
cvs(svm.SVC(kernel='rbf' , C=10, gamma='auto'),x_train, y_train,cv=5)
```

```python
cvs(svm.SVC(kernel='rbf' , C=20, gamma='auto'),x_train, y_train,cv=5)
```

```python
kernels = ['rbf' , 'linear']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cvs(svm.SVC(kernel=kval, C=cval , gamma='auto'),x_train , y_train,cv=5)
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)
        
         
avg_scores      
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)

clf.fit(x_train , y_train)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
```

```python
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

## RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rs = RandomizedSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},
    cv=5,
    return_train_score=False,
    n_iter=2
                      )
```

```python
rs.fit(iris.data,iris.target)
```

```python
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```
